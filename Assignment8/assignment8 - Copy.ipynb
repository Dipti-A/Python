```python
import pandas as pd
from datetime import date
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
# This function is created for data cleaning as the dates are in different format. Moreover where year is not specified in yyyy
# format if the value is less than 69 then it is read as 2069 instead of 1969 hence below method takes care of fixing this issue

def fixDate(x):

    if x.year > 2020:
        year = x.year - 100
    else:
        year = x.year
    #print(year,x.month,x.day)
    return datetime(year,x.month,x.day)

```

```python
df = pd.read_csv("HRTest.csv")

#df.describe()
#df.info()
df.DOB = pd.to_datetime(df.DOB)

df['DOB'] = df['DOB'].apply(fixDate)
df.head(10)
                           
```

```python
# Reference: https://www.geeksforgeeks.org/python-program-to-calculate-age-in-year/
#2. Calculate Age of all the employees

def calculateAge(dob): 
    today = date.today() 
    age = today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))   
    return age

df=df.assign(Age = df['DOB'].apply(calculateAge))
#df['Age'] = df['DOB'].apply(calculateAge)
df.head(10)
```
